# Лабораторна робота №2

## Дослідження закону великих чисел

Мета роботи: дослідити поведінку середнього арифметичного
Sn / n для послідовності незалежних однаково розподілених випадкових величин.

## 1. Постановка задачі

Розглядається послідовність незалежних однаково розподілених випадкових величин:

Xj = 1 + Yj,

де

Yj ~ Beta(4,1)

Потрібно:

1. Знайти щільність розподілу f(t).
2. Обчислити E[X] та Var[X].
3. Перевірити виконання закону великих чисел.
4. Побудувати графіки Sn/n.
5. Провести симуляції для різних n.
6. Побудувати гістограми та зробити висновки.

## 2. Теоретичні викладки

Оскільки:

Y ~ Beta(4,1),

то щільність Beta-розподілу:

fY(y) = 4y^3, 0 < y < 1

Випадкова величина:

X = 1 + Y

тоді:

1 < X < 2

Щільність X:

f(x) = 4(x - 1)^3, 1 < x < 2

---

Для Beta(4,1):

E[Y] = 4 / (4+1) = 4/5 = 0.8

Var[Y] = (4·1) / ((5^2)(6))
       = 4 / 150
       = 2/75

---

Оскільки X = 1 + Y:

E[X] = 1 + E[Y]
     = 1.8

Var[X] = Var[Y]
       = 2/75 ≈ 0.02667

---

За законом великих чисел:

Sn/n → E[X] = 1.8

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 3. Генерація випадкових величин

Для моделювання використовуємо розподіл Beta(4,1).

```python
def generate_X(n):
    y = np.random.beta(4, 1, size=n)
    x = 1 + y
    return x
```

```python
x = generate_X(100000)

print("Емпіричне E[X] =", np.mean(x))
print("Емпіричне Var[X] =", np.var(x))
```

## 4. Перевірка закону великих чисел

Будуємо графіки:

Sn/n

для 10 симуляцій.

```python
n = 1000

plt.figure(figsize=(10,6))

for _ in range(10):
    x = generate_X(n)

    S = np.cumsum(x)
    avg = S / np.arange(1, n+1)

    plt.plot(avg)

plt.axhline(y=1.8, color='red', linestyle='--')
plt.title("Збіжність Sn/n")
plt.xlabel("n")
plt.ylabel("Sn/n")

plt.show()
```

## 5. Симуляції

Нехай B = 1000.

Для кожного n:
- генеруємо B значень Sn/n;
- обчислюємо середнє та дисперсію;
- оцінюємо ймовірність:

P(|Sn/n - 1.8| ≥ ε)

де ε = 0.001

```python
B = 1000
eps = 0.001

ns = [10, 100, 1000, 10000]

results = {}

for n in ns:

    values = []

    for _ in range(B):

        x = generate_X(n)

        values.append(np.mean(x))

    values = np.array(values)

    mean_val = np.mean(values)
    var_val = np.var(values)

    prob = np.mean(np.abs(values - 1.8) >= eps)

    results[n] = values

    print(f"\nn = {n}")
    print(f"E = {mean_val:.6f}")
    print(f"Var = {var_val:.8f}")
    print(f"P(|Sn/n - 1.8| >= {eps}) = {prob:.6f}")
```

## 6. Гістограми

Будуємо гістограми розподілу Sn/n.

```python
for n in ns:

    plt.figure(figsize=(7,4))

    plt.hist(results[n],
             bins=30,
             edgecolor='black')

    plt.axvline(x=1.8,
                color='red',
                linestyle='--')

    plt.title(f"Гістограма Sn/n, n={n}")

    plt.show()
```

## 7. Висновки

У роботі було досліджено поведінку середнього арифметичного:

Sn/n

для випадкових величин Xj = 1 + Yj, де Yj ~ Beta(4,1).

Було встановлено:

- емпіричні характеристики наближаються до теоретичних;
- зі збільшенням n дисперсія Sn/n зменшується;
- Sn/n збігається до E[X] = 1.8;
- ймовірність великих відхилень зменшується.

Отримані результати підтверджують закон великих чисел.

# Частина 2. Центральна гранична теорема

Розглядаємо:

S̃n = (Sn − E[Sn]) / √n

Досліджуємо виконання центральної граничної теореми.

## 1. Теоретичні викладки

Для випадкових величин Xj:

E[X] = 1.8

Var[X] = 2/75 ≈ 0.02667

Оскільки:
- Xj незалежні,
- однаково розподілені,
- мають скінченне математичне сподівання та дисперсію,

то виконується центральна гранична теорема.

---

Розглядаємо:

S̃n = (Sn − E[Sn]) / √n

де:

Sn = X1 + X2 + ... + Xn

---

За центральною граничною теоремою:

S̃n ⇒ N(0, σ²)

де:

σ² = Var[X] = 2/75

Отже, слабкою границею є нормальний розподіл:

N(0, 2/75)

```python
from scipy.stats import norm
```

## 2. Генерація випадкових величин S̃n

```python
EX = 1.8
VAR = 2 / 75

def generate_S_tilde(n):

    x = generate_X(n)

    Sn = np.sum(x)

    S_tilde = (Sn - n * EX) / np.sqrt(n)

    return S_tilde
```

## 3. Симуляції

Нехай B = 1000.

Для кожного n:
- генеруємо B значень S̃n;
- обчислюємо емпіричне середнє;
- обчислюємо емпіричну дисперсію.

```python
B = 1000

ns = [10, 100, 1000, 10000]

S_tilde_results = {}

for n in ns:

    values = np.array([generate_S_tilde(n) for _ in range(B)])

    S_tilde_results[n] = values

    print(f"\nn = {n}")
    print(f"Емпіричне середнє = {np.mean(values):.5f}")
    print(f"Емпірична дисперсія = {np.var(values):.5f}")
```

## 4. Емпірична функція розподілу

Порівнюємо емпіричну функцію розподілу з теоретичною нормальною функцією розподілу.

```python
for n in ns:

    data = np.sort(S_tilde_results[n])

    ecdf = np.arange(1, len(data)+1) / len(data)

    plt.figure(figsize=(7,4))

    plt.plot(data, ecdf, label='Емпірична F(x)')

    x = np.linspace(min(data), max(data), 500)

    plt.plot(
        x,
        norm.cdf(x, loc=0, scale=np.sqrt(VAR)),
        label='Гранична N(0, 2/75)'
    )

    plt.title(f"Функція розподілу S̃n, n={n}")

    plt.legend()

    plt.show()
```

## 5. Гістограми відносних частот

Порівнюємо гістограму з щільністю нормального розподілу.

```python
for n in ns:

    data = S_tilde_results[n]

    plt.figure(figsize=(7,4))

    plt.hist(
        data,
        bins=30,
        density=True,
        edgecolor='black',
        alpha=0.7
    )

    x = np.linspace(min(data), max(data), 500)

    plt.plot(
        x,
        norm.pdf(x, loc=0, scale=np.sqrt(VAR))
    )

    plt.title(f"Гістограма S̃n, n={n}")

    plt.show()
```

## 6. Висновки

У ході роботи було досліджено виконання центральної граничної теореми.

Було встановлено:

- емпіричні розподіли S̃n наближаються до нормального;
- емпіричні функції розподілу збігаються до теоретичної функції нормального розподілу;
- гістограми набувають форми дзвону;
- емпіричні характеристики стабілізуються.

Отримані результати підтверджують виконання центральної граничної теореми.
