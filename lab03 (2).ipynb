### 1. Властивості циклічних груп

Для $n=100,173,530,746,911,1024$ розглянемо циклічну групу $\mathbb{Z}_n$ (або $C_n$):
1. Для кожного дільника $d$ порядку $n$:<br>
   1.1 Знайдіть кількість елементів порядку $d$ в групі $\mathbb{Z}_n$ (або $C_n$).<br>
   1.2 Порівняйте з $\phi(d)$.
2. Знайдіть кількість підгруп в групі $\mathbb{Z}_n$ (або $C_n$).<br>
   Порівняйте з $\tau(n)$.

Проаналізуйте результати та дайте відповіді на запитання:
1. Скільки елементів порядку $k$ в циклічній групі порядку $n$?
2. Cкільки підгруп в циклічній групі порядку $n$?

```python
print("Аналіз властивостей циклічних груп Z_n")
print("=" * 50)

n_values = [100, 173, 530, 746, 911, 1024]

for n in n_values:
    print(f"\n\nАналіз для n = {n}:")

    
    divisors_n = divisors(n)
    print(f"Дільники n: {divisors_n}")
    
    subgroup_count = len(divisors_n)
    print(f"Кількість підгруп: {subgroup_count} (τ(n) = {subgroup_count})")
    
    print(f"\nЕлементи порядку d:")
    for d in divisors_n:
        elements_of_order_d = euler_phi(d)
        print(f"  d = {d}: {elements_of_order_d} елемент(ів) (φ({d}) = {elements_of_order_d})")

print("ЗАГАЛЬНІ ВЛАСТИВОСТІ ЦИКЛІЧНИХ ГРУП:")


demo_n = 12
print(f"\nДемонстрація для n = {demo_n}:")
G = CyclicPermutationGroup(demo_n) 
divs = divisors(demo_n)

print(f"Дільники {demo_n}: {divs}")
print(f"Кількість підгруп: {len(divs)}")

print(f"\nЕлементи різних порядків:")
for d in divs:
    phi_d = euler_phi(d)
    print(f"  Порядок {d}: {phi_d} елемент(ів)")

print(f"\nВИСНОВКИ:")
print("1. Кількість елементів порядку k в циклічній групі порядку n:")
print("   - Якщо k | n, то кількість дорівнює φ(k)")
print("   - Якщо k ∤ n, то таких елементів немає")
print(f"2. Кількість підгруп в циклічній групі порядку n дорівнює τ(n) = {len(divs)}")
```

### 2. Порядки елементів в дієдральних групах

Для $n=3,4,...,20$ розглянемо дієдральну групу $D_n$:   
1. Знайдіть кількість елементів порядку $d=2$ в групі $D_n$.
2. Для кожного дільника $d>2$ порядку $|D_n|=2n$:<br>
   Знайдіть кількість елементів порядку $d$ в групі $D_n$.

Проаналізуйте результати та дайте відповіді на запитання:
1. Скільки елементів порядку $d=2$ в дієдральній групі $D_n$?
2. Скільки елементів порядку $d$ в дієдральній групі $D_n$?
3. Чи для кожного дільника $d<2n$ порядку групи $D_n$ існує елемент порядку $d$?

```python
print("Аналіз порядків елементів у дієдральних групах D_n")


for n in range(3, 21):
    D = DihedralGroup(n)
    
    print(f"\nДієдральна група D_{n} (порядок {2*n}):")

    
    elements_order_2 = [g for g in D if g.order() == 2]
    count_order_2 = len(elements_order_2)
    
    print(f"Елементи порядку 2: {count_order_2}")
    
    print(f"\nЕлементи порядків d > 2:")
    found_orders = set()
    
    for d in divisors(2*n):
        if d <= 2:
            continue
            
        elements_d = [g for g in D if g.order() == d]
        count_d = len(elements_d)
        
        if count_d > 0:
            found_orders.add(d)
            print(f"  Порядок {d}: {count_d} елемент(ів)")
    
    all_divisors = set(divisors(2*n))
    all_divisors.discard(2*n)  
    missing_orders = all_divisors - found_orders - {1}  
    
    if missing_orders:
        print(f"Не існує елементів порядків: {sorted(missing_orders)}")
    else:
        print("Для всіх дільників d < 2n існують елементи порядку d")


print("ЗАГАЛЬНІ ВЛАСТИВОСТІ ДІЄДРАЛЬНИХ ГРУП:")

print("\n1. Елементи порядку 2 в D_n:")
print("   - Якщо n непарне: n елементів порядку 2")
print("   - Якщо n парне: n + 1 елемент порядку 2")

print("\n2. Елементи порядку d > 2 в D_n:")
print("   - Можливі порядки: дільники n, що більші за 2")
print("   - Кількість елементів порядку d (де d|n та d>2): φ(d)")

print("\n3. Існування елементів заданого порядку:")
print("   - Елемент порядку d існує тоді і тільки тоді, коли:")
print("     * d = 2 (завжди існує)")
print("     * d ділить n та d > 2")
print("   - Не існує елементів порядку, які не є дільниками n або 2")

print("ДЕМОНСТРАЦІЯ ДЛЯ D_6:")


D6 = DihedralGroup(6)
print(f"Порядок групи D_6: {D6.order()}")

orders_count = {}
for g in D6:
    order_g = g.order()
    orders_count[order_g] = orders_count.get(order_g, 0) + 1

print("Розподіл елементів за порядками:")
for order_val, count in sorted(orders_count.items()):
    print(f"  Порядок {order_val}: {count} елемент(ів)")

print("\nТеоретичний розрахунок для D_6:")
print(f"Елементи порядку 2: 7 (6 відображень + 1 обертання 180°)")
print(f"Елементи порядку 3: φ(3) = 2")
print(f"Елементи порядку 6: φ(6) = 2")
```

### 3. Дискретний логарифм в групі $\mathbb{Z}_n^{*}$

Розглянемо групу $\mathbb{Z}_n^{*}$ для $n=1102158078129785185997827239806$.
1. Перевірте, що група є циклічною і знайдіть її порядок.
2. Знайдіть найменше натуральне $a\in\mathbb{N}$ таке, що елемент $\overline{a}\in\mathbb{Z}_n^{*}$ є твірним групи.
3. Перевірте, що порядок елемента $\overline{a}$ дорівнює $\phi(n)$.
4. Виберіть три випадкових елементи $g_1,g_2,g_3$ групи $\mathbb{Z}_n^{*}$.
5. Для кожного $g$ знайдіть степінь $k$ такий, що $g=\overline{a}^k$.


```python
n = 1102158078129785185997827239806

if n % 2 == 0:
    p = n // 2
    if is_prime(p):
        print(f"n = 2 * p, де p = {p} є простим.")
        phi_n = p - 1
        print(f"Порядок групи Z_n*: φ(n) = {phi_n}")
        
        a_mod_p = primitive_root(p)
        print(f"Найменший примітивний корінь modulo p: {a_mod_p}")
        
        if a_mod_p % 2 == 1:
            a = a_mod_p
        else:
            a = a_mod_p + p
        print(f"Твірний елемент a в Z_n*: {a}")
        
        order_a = Mod(a, n).multiplicative_order()
        print(f"Порядок елемента a: {order_a}")
        if order_a == phi_n:
            print("Порядок a дорівнює φ(n).")
        else:
            print("Помилка: порядок a не дорівнює φ(n).")
        
        random_elements = []
        for i in range(3):
            while True:
                g = randint(2, n-1)
                if gcd(g, n) == 1:
                    random_elements.append(g)
                    break
        print("\nВипадкові елементи g1, g2, g3:")
        for i, g in enumerate(random_elements):
            print(f"g_{i+1} = {g}")
        
        print("\nЗнаходження k таких, що g = a^k mod n:")
        for i, g in enumerate(random_elements):
            k = discrete_log(Mod(g, n), Mod(a, n))
            print(f"Для g_{i+1}: k = {k}")
            if Mod(a, n)^k == Mod(g, n):
                print(f"Перевірка для g_{i+1} пройшла успішно.")
            else:
                print(f"Помилка перевірки для g_{i+1}.")
    else:
        print("p не є простим, тому група Z_n* не циклічна.")
else:
    print("n не парне, тому група Z_n* не циклічна.")
```
