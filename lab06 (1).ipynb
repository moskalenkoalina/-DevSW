### 1. Розширення полів

1. Побудуйте розширення $K=\mathbb{Q}(\sqrt[3]{2},i)$ поля $\mathbb{Q}$.
2. Знайдіть степінь розширення $[K:\mathbb{Q}]$ та базис $K$ над $\mathbb{Q}$.
3. Задайте елемент $a=\sqrt[3]{2}+i$ і знайдіть його мінімальний многочлен $m_a(x)$ над полем $\mathbb{Q}$.
4. Побудуйте розширення $L=\mathbb{Q}(\sqrt[3]{2}+i)$ як факторкільце $\mathbb{Q}[x]/(m_a(x))$.
5. Перевірте, чи є поля $K$ та $L$ рівними, порівнявши степені розширення.



```python
# 1. Задаємо елемент  та знаходимо його мінімальний многочлен
print("=== Елемент  та його мінімальний многочлен ===")

# Будуємо поле  як абсолютне розширення
R.<x> = QQ[]
K_abs.<a, b> = NumberField([x^3 - 2, x^2 + 1])

# Задаємо елемент a
a_elem = a + b
print(f"Елемент a = {a_elem}")
print(f"a = {a} + {b}")

# Знаходимо мінімальний многочлен a над Q
m_a = a_elem.absolute_minpoly()
print(f"\nМінімальний многочлен m_a(x) = {m_a}")

# Перевіряємо, що a дійсно є коренем цього многочлена
print(f"Перевірка: m_a(a_elem) = {m_a(a_elem)} (має бути 0)")

# 2. Будуємо розширення L  як факторкільце Q[x]/(m_a(x))
print("\n\n=== Побудова розширення L  ===")

L.<alpha> = NumberField(m_a)
print(f"L = Q[x]/({m_a})")
print(f"Ступінь розширення [L:Q] = {L.absolute_degree()}")

# Базис L над Q
print(f"Базис L над Q: {[alpha^i for i in range(L.absolute_degree())]}")

# 3. Перевіряємо, чи є поля K та L рівними
print("\n\n=== Порівняння полів K та L ===")
print(f"K = ")
print(f"L = )")
print(f"Ступінь [K:Q] = {K_abs.absolute_degree()}")
print(f"Ступінь [L:Q] = {L.absolute_degree()}")

# Порівнюємо ступені розширення
if K_abs.absolute_degree() == L.absolute_degree():
    print("\nСтупені розширень рівні! [K:Q] = [L:Q]")
    print("Це означає, що L може бути ізоморфним K.")
    print("Але для точної перевірки рівності потрібно додатково дослідити, чи є K і L ізоморфними.")
else:
    print("\nСтупені розширень не рівні!")
    print("Це означає, що поля K і L не можуть бути рівними.")
```

### 2. Поля розкладу

1. Побудуйте поле розкладу $L$ многочлена $f(x)=x^4+x^2+1$ над полем $\mathbb{Q}$.
2. Знайдіть степінь та базис $L$ над $\mathbb{Q}$.
3. Розкладіть многочлен $f(x)$ на лінійні множники над полем $L$.
4. Побудуйте поле розкладу $K$ многочлена $g(x)=x^5+3x^2+2$ над полем $\mathbb{Z}_3$.
5. Знайдіть степінь та базис $K$ над $\mathbb{Z}_3$.
6. Розкладіть многочлен $g(x)$ на лінійні множники над полем $K$.

```python
print("=== Частина 1: Поле розкладу над Q ===")
print()

# 1. Побудова поля розкладу многочлена f(x) = x^4 + x^2 + 1 над Q
print("1. Побудова поля розкладу L многочлена f(x) = x^4 + x^2 + 1 над Q")

# Многочлен
x = polygen(QQ)
f = x^4 + x^2 + 1
print(f"   f(x) = {f}")

# Відомо алгебраїчно, що f(x) = (x^2 + x + 1)(x^2 - x + 1)
print(f"\n   Алгебраїчний факт: f(x) = (x^2 + x + 1)(x^2 - x + 1)")
print(f"   Перевірка множенням:")
print(f"   (x^2 + x + 1)(x^2 - x + 1) = {(x^2 + x + 1)*(x^2 - x + 1)}")

# Корені першого множника: (-1 ± √(-3))/2
# Корені другого множника: (1 ± √(-3))/2
print(f"\n   Корені x^2 + x + 1: (-1 ± √(-3))/2")
print(f"   Корені x^2 - x + 1: (1 ± √(-3))/2")
print(f"   Усі корені містять √(-3)")

# Будуємо поле Q(√(-3))
print(f"\n   Будуємо поле Q(√(-3)):")

try:
    # Будуємо поле як NumberField
    L.<a> = NumberField(x^2 + 3)
    print(f"   L = Q(a), де a^2 + 3 = 0")
    print(f"   Тобто a = √(-3) = i√3")
    print(f"   L = {L}")
    
    # Ступінь розширення
    print(f"\n2. Ступінь розширення L над Q:")
    print(f"   [L : Q] = {L.absolute_degree()}")
    
    # Для NumberField можемо отримати базис
    try:
        basis_L = L.basis()
        print(f"   Базис L над Q: {basis_L}")
    except:
        print(f"   Базис L над Q: {{1, a}} де a = √(-3)")
    
except Exception as e:
    print(f"   Помилка: {e}")
    print(f"   Використовуємо символьний підхід:")
    print(f"   Поле розкладу: Q(√(-3))")
    print(f"   Ступінь: 2")
    print(f"   Базис: {{1, √(-3)}}")

print(f"\n3. Розклад f(x) на лінійні множники над L:")
print(f"   Якщо a = √(-3), то:")
print(f"   Корені: (-1 ± a)/2 і (1 ± a)/2")
print(f"   Отже:")
print(f"   f(x) = (x - (-1+a)/2)(x - (-1-a)/2)(x - (1+a)/2)(x - (1-a)/2)")
print(f"   = (x^2 + x + 1)(x^2 - x + 1)")

print()

print("=== Частина 2: Поле розкладу над Z3 ===")
print()

# 4. Побудова поля розкладу многочлена g(x) = x^5 + 3x^2 + 2 над Z3
print("4. Побудова поля розкладу K многочлена g(x) = x^5 + 3x^2 + 2 над Z3")

# Поле Z3
Z3 = GF(3)
print(f"   Z3 = {Z3}")
print(f"   Елементи Z3: {[el for el in Z3]}")

# Многочлен g(x) = x^5 + 3x^2 + 2
# Оскільки 3 = 0 в Z3, то g(x) = x^5 + 2
x_z3 = polygen(Z3)
g = x_z3^5 + 3*x_z3^2 + 2
print(f"\n   У Z3: 3 = 0, тому g(x) = {g}")
print(f"   Тобто g(x) = x^5 + 2 (бо 3*x^2 = 0)")

# Перевірка коренів у Z3
print(f"\n   Перевірка коренів у Z3:")
for el in Z3:
    value = el^5 + 2
    print(f"   g({el}) = {el}^5 + 2 = {value}")
    if value == 0:
        print(f"     → {el} є коренем у Z3")

# x=1 є коренем, бо 1^5 + 2 = 1 + 2 = 3 ≡ 0 mod 3
print(f"\n   Отже, x=1 є коренем, тому (x-1) ділить g(x)")

# Виконуємо ділення многочленів
quotient = g // (x_z3 - 1)
remainder = g % (x_z3 - 1)
print(f"   g(x) / (x-1) = {quotient}")
print(f"   Залишок: {remainder} (має бути 0)")

print(f"\n   Отже, g(x) = (x-1)({quotient})")

# Дослідимо многочлен-частку
print(f"\n   Досліджуємо частку {quotient}:")
print(f"   Ступінь: {quotient.degree()}")

# Перевіримо, чи має він корені у Z3
print(f"   Перевірка коренів частки у Z3:")
has_roots_in_z3 = False
for el in Z3:
    value = quotient(el)
    print(f"   q({el}) = {value}")
    if value == 0:
        print(f"     → {el} є коренем частки у Z3")
        has_roots_in_z3 = True

if not has_roots_in_z3:
    print(f"   Частка не має коренів у Z3")
    print(f"   Отже, вона незвідна над Z3 або розкладається на множники вищого степеня")

print(f"\n   Для знаходження поля розкладу потрібно розширити Z3")
print(f"   Розглянемо розширення найменшого степеня, що містить всі корені")

# Побудуємо розширення через корінь частки
# Оскільки частка має ступінь 4, мінімальне розширення має ступінь, що ділить 4
print(f"\n   Частка має ступінь 4, тому розширення має ступінь, що ділить 4")
print(f"   Найменше розширення - це поле розкладу цього многочлена 4-го степеня")

# Спростимо: візьмемо GF(3^4) як можливе поле розкладу
n = 4
K = GF(3^n, 'b')
field_size = 3^n
print(f"\n   Візьмемо GF(3^{n}) як поле розкладу")
print(f"   K = GF({field_size}) з {field_size} елементами")
print(f"   Примітивний елемент: b")

# 5. Ступінь та базис K над Z3
print(f"\n5. Ступінь та базис K над Z3")
print(f"   [K : Z3] = {K.degree()}")

# Для скінченних полів отримуємо векторний простір і його базис
# K.vector_space() повертає кортеж (V, from_V, to_V)
V, from_V, to_V = K.vector_space()
# Базис векторного простору над Z3
basis_vectors = V.basis()
# Конвертуємо вектори в елементи поля K
basis_elements = [from_V(v) for v in basis_vectors]
print(f"   Базис K над Z3: {basis_elements}")



# 6. Розклад g(x) на лінійні множники над K
print(f"\n6. Розклад g(x) на лінійні множники над K")

# Поліном над K
x_K = polygen(K)
g_K = x_K^5 + 2

print(f"   У полі K многочлен g(x) = x^5 + 2")
print(f"   Знайдемо всі корені в K:")

# Оскільки K має 81 елемент, ми можемо знайти корені перебором
roots = []
for i, element in enumerate(K):
    if g_K(element) == 0:
        roots.append(element)
    # Для швидкості можна зупинитись, коли знайдемо 5 коренів
    if len(roots) == 5:
        break

print(f"   Знайдено {len(roots)} коренів:")
for i, root in enumerate(roots):
    print(f"     Корінь {i+1}: {root}")

print(f"\n   Розклад на лінійні множники:")
if len(roots) == 5:
    print(f"   g(x) = (x - {roots[0]})(x - {roots[1]})(x - {roots[2]})(x - {roots[3]})(x - {roots[4]})")
    print(f"   (Ураховуючи, що корені можуть бути кратними)")
else:
    print(f"   Знайдено тільки {len(roots)} коренів, але очікується 5.")
    print(f"   Це може бути через кратні корені або те, що не всі корені в K.")
    print(f"   Можливо, потрібне більше розширення.")

print(f"\n   Перевірка: чи справді всі 5 коренів належать K?")
print(f"   Оскільки K = GF(81) і x^81 - x містить всі многочлени степеня, що ділить 80,")
print(f"   а 5 ділить 80, то многочлен x^5 + 2 повністю розкладається в K.")

print("\n=== Висновки ===")
print("1. Для f(x) = x^4 + x^2 + 1:")
print("   - Поле розкладу: Q(√(-3))")
print("   - Ступінь: 2")
print("   - Базис: {1, √(-3)}")
print("   - Розклад: (x - (-1+√(-3))/2)(x - (-1-√(-3))/2)(x - (1+√(-3))/2)(x - (1-√(-3))/2)")
print()
print("2. Для g(x) = x^5 + 3x^2 + 2 над Z3:")
print("   - Поле розкладу: GF(3^4) = GF(81)")
print("   - Ступінь: 4")
print("   - Базис: {1, b, b^2, b^3} де b - примітивний елемент GF(81)")
print("   - g(x) розкладається на 5 лінійних множників у GF(81)")
```

### 3. Скінченні поля

1. Побудуйте поле $F$ з $8$ елементів. Виведіть таблицю множення.
2. Побудуйте факторкільця $K=\mathbb{Z}_2[x]/(x^3+x+1)$ та $L=\mathbb{Z}_2[x]/(x^3+x^2+1)$.
3. Перевірте, що $K$ та $L$ є полями з $8$ елементів.
4. Знайдіть первісні елементи для полів $K$ та $L$.
5. Встановіть ізоморфізм між полями $K$ та $L$.


```python

# 1. Поле F з 8 елементів та таблиця множення


# Створюємо поле GF(2^3) з примітивним елементом a
F.<a> = GF(2^3, name='a')  
print("Поле F = GF(2^3) з примітивним елементом a")
print("Елементи поля F:", list(F))

# Таблиця множення
print("\nТаблиця множення поля F:")
elements = list(F)

# Заголовок
print("    ", end="")
for e in elements:
    print(f"{str(e):>3}", end="")
print()

# Рядки таблиці
for x in elements:
    print(f"{str(x):>3}", end="")
    for y in elements:
        print(f"{str(x*y):>3}", end="")
    print()

# 2. Факторкільця K та L


R.<x> = PolynomialRing(GF(2))

# Поле K = Z2[x]/(x^3 + x + 1)
fK = x^3 + x + 1
K.<k> = GF(2^3, modulus=fK)
print("\nПоле K = Z2[x]/(x^3 + x + 1)")
print("Елементи K:", list(K))

# Поле L = Z2[x]/(x^3 + x^2 + 1)
fL = x^3 + x^2 + 1
L.<l> = GF(2^3, modulus=fL)
print("\nПоле L = Z2[x]/(x^3 + x^2 + 1)")
print("Елементи L:", list(L))

# 3. Перевірка, що це поля
print("\nПеревірка властивостей полів:")
print("K є полем?", K.is_field())
print("L є полем?", L.is_field())

########################################
# 4. Первісні елементи (генератори)
########################################

def find_primitive_elements(F):
    prims = []
    for e in F:
        if e != 0 and e.multiplicative_order() == F.order() - 1:
            prims.append(e)
    return prims

primK = find_primitive_elements(K)
primL = find_primitive_elements(L)

print("\nПервісні елементи K:")
for e in primK:
    print(e, "є первісним елементом K")

print("\nПервісні елементи L:")
for e in primL:
    print(e, "є первісним елементом L")

# Візьмемо перший первісний елемент для ізоморфізму
genK = primK[0]
genL = primL[0]
print(f"\nВибрані генератори: genK = {genK}, genL = {genL}")

########################################
# 5. Ізоморфізм K → L вручну
########################################

def iso_phi(elem):
    if elem == 0:
        return L(0)
    # Знаходимо степінь генератора genK
    for i in range(7):
        if elem == genK^i:
            return genL^i
    return None  # на випадок помилки

print("\nІзоморфізм φ: K → L")
for elem in K:
    print(f"φ({elem}) = {iso_phi(elem)}")


```
