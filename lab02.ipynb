### 4. Системи твірних

Для кожної з груп $\mathbb{Z}_{24}$, $Q_8$, $D_5$, $S_4$ виконайте наступні кроки:

1. Створіть групу.
2. Знайдіть деяку систему твірних групи.    % G.gens()
3. Знайдіть всі мінімальні системи твірних групи.    % G.minimal_generating_sets()
4. Візьміть довільні два елементи групи, породіть ними підгрупу, знайдіть її порядок.  % G.subgroup([el1, el2])

```python

G = CyclicPermutationGroup(24)
print(f"1. Група створена, порядок: {G.order()}")
print("\n2. Система твірних групи (G.gens()):")
generators = G.gens()
print(f"   Твірні: {generators}")
print(f"   Кількість твірних: {len(generators)}")
print("\n3. Мінімальні системи твірних:")
try:
    minimal_sets = list(G.minimal_generating_sets())
    print(f"   Кількість мінімальних систем: {len(minimal_sets)}")
    for i, gen_set in enumerate(minimal_sets[:3]): 
        print(f"   Система {i+1}: {list(gen_set)}")
except:
    print("   Метод minimal_generating_sets() не підтримується")
    elements = list(G)
    for g in elements:
        if g.order() == G.order():
            print(f"   Мінімальна система: [{g}]")
            break
print("\n4. Підгрупа, породжена двома довільними елементами:")
elements = list(G)
a, b = elements[1], elements[2]  
print(f"   Елемент a: {a} (порядок: {a.order()})")
print(f"   Елемент b: {b} (порядок: {b.order()})")

H = G.subgroup([a, b])
print(f"   Підгрупа H = <a, b>")
print(f"   Порядок підгрупи H: {H.order()}")

if H.order() == G.order():
    print("   Результат: H = G (підгрупа збігається з усією групою)")
elif H.order() == 1:
    print("   Результат: H - тривіальна підгрупа")
else:
    print("   Результат: H - власна підгрупа")
```
